```python
# Import matplotlib to display images
import matplotlib.pyplot as plt
%matplotlib inline
# Set the default size of the figures to be 
# width 18.0 and height 15.0 units.
plt.rcParams['figure.figsize'] = (18.0, 13.0)

# Import NumPy in order to execute array operations efficiently.
import numpy as np

# Enable the OpenCV 3 module. The module name is not changed to 
# cv3 yet which could be quite confusing sometimes.
import cv2

# This is a helper function which displays an OpenCV image
def display(img):
    # OpenCV stores images in B G R sequence, so we need
    # to explicitly convert it to R G B in order to plot
    # with matplotlib
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
```

```python
img = cv2.imread("./imgs/segmentation/nao.jpg")
display(img)
```

```python
rgb_pts = np.float32(img.reshape(-1, 3))

K = 16
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)

ret,labels,centers=cv2.kmeans(rgb_pts, K, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

# Now convert back into uint8, and make original image
centers = np.uint8(centers)

seg_rgb_pts = centers[labels.flatten()]
seg_rgb_img = seg_rgb_pts.reshape((img.shape))

plt.subplot(1, 2, 1)
display(seg_rgb_img)
plt.subplot(1, 2, 2)
display(img)
```

Joint space-color hyperspace


```python
xy_weight = 0.2 # 0.2, 0.5, 1.0
xy_pts = xy_weight * np.indices(img.shape[:2]).reshape(2, -1).transpose()
rgbxy_pts = np.float32(np.hstack((rgb_pts, xy_pts)))

K = 16
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
ret,labels,centers=cv2.kmeans(rgbxy_pts, K, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

# Now convert back into uint8, and make original image
centers = np.uint8(centers[:, :3])

seg_rgbxy_pts = centers[labels.flatten()]
seg_rgbxy_img = seg_rgbxy_pts.reshape((img.shape))

plt.subplot(1, 2, 1)
plt.title("RGBXY space")
display(seg_rgbxy_img)
plt.subplot(1, 2, 2)
plt.title("RGB space")
display(seg_rgb_img)
```

```python
xy_r = 20
rgb_r = 80
meanshift_img = cv2.pyrMeanShiftFiltering(img, xy_r, rgb_r)
display(meanshift_img)

plt.subplot(1, 2, 1)
display(meanshift_img)
plt.subplot(1, 2, 2)
display(img)
```

```python
fg_roi = (250, 550, 380, 500)

x, y, w, h = fg_roi
display(cv2.rectangle(img.copy(), (x, y), (x + w, y + h), (0, 255, 255), 3)) 
```

```python
mask = np.zeros(img.shape[:2], np.uint8)

bg_model = np.zeros((1, 65), np.float64)
fg_model = np.zeros((1, 65), np.float64)
cv2.grabCut(img, mask, fg_roi, bg_model, fg_model, 5, cv2.GC_INIT_WITH_RECT)

mask = np.where((mask == cv2.GC_FGD) | (mask == cv2.GC_PR_FGD), 1, 0).astype('uint8')
grab_cut_img = img * mask[:, :, np.newaxis]


display(grab_cut_img)
```

```python
mask_img = cv2.imread("./imgs/segmentation/mask.png")

mask = cv2.cvtColor(mask_img, cv2.COLOR_BGR2GRAY)
mask[mask == 0] = cv2.GC_BGD
mask[mask == 128] = cv2.GC_PR_BGD
mask[mask == 255] = cv2.GC_FGD

bg_model = np.zeros((1, 65), np.float64)
fg_model = np.zeros((1, 65), np.float64)
cv2.grabCut(img, mask, None, bg_model, fg_model, 5, cv2.GC_INIT_WITH_MASK)

mask = np.where((mask == cv2.GC_FGD) | (mask == cv2.GC_PR_FGD), 1, 0).astype('uint8')
grab_cut_img = img * mask[:, :, np.newaxis]

plt.subplot(1, 2, 1)
display(grab_cut_img)
plt.subplot(1, 2, 2)
display(mask_img)
```
