# Image Segmentation

As before we start with setting up the environment and a simple helper function to display images.

```python
# Import matplotlib to display images
import matplotlib.pyplot as plt
%matplotlib inline
# Set the default size of the figures to be 
# width 18.0 and height 15.0 units.
plt.rcParams['figure.figsize'] = (18.0, 13.0)

# Import NumPy in order to execute array operations efficiently.
import numpy as np

# Enable the OpenCV 3 module. The module name is not changed to 
# cv3 yet which could be quite confusing sometimes.
import cv2

# This is a helper function which displays an OpenCV image
def display(img):
    # OpenCV stores images in B G R sequence, so we need
    # to explicitly convert it to R G B in order to plot
    # with matplotlib
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
```

Load the image which we will be working with:

```python
img = cv2.imread("./imgs/segmentation/nao.jpg")
display(img)
```

## K-means Clustering Segmentation 

```python
# Construct an array of 3D points in the RGB colour space.
# Each pixel in the image is represented as a single point.
rgb_pts = np.float32(img.reshape(-1, 3))

# Set the number of clusters to 16
K = 16
# Create the criteria when the K-means iterations should be terminated. 
# The iteration will stop if the error is less then EPS = 1.0 or the
# maximum number of iteration MAX_ITER = 20 is achieved
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 20, 1.0)

# Run K-means. The return values are:
# ret     - the final value of the minimised sum of distances
# labels  - the cluster which each point belons to
# centers - the centers of the K clusters
ret, labels, centers=cv2.kmeans(rgb_pts,  # Input points
                                K,        # Number of clusters 
                                None,     # Initial estimate of the label of each point
                                criteria, # Termination criteria
                                10,       # Return the best result after that many attempts
                                cv2.KMEANS_RANDOM_CENTERS) # Initialise the clusters randomly

# Convert the centers from float32 to uint8
centers = np.uint8(centers)

# Set the coordinates of each point to the center of the cluster it belongs to
seg_rgb_pts = centers[labels.flatten()]
# Make the RGB points array into an image again
seg_rgb_img = seg_rgb_pts.reshape((img.shape))

# Show the original and the segmented image
plt.subplot(1, 2, 1)
plt.title('Original image')
display(img)
plt.subplot(1, 2, 2)
plt.title('Segmented image')
display(seg_rgb_img)

```

## Joint Space-Colour Hyperspace Segmentation

A straight forward extension of the K-Means segmentation algorithm is to represent the pixels in a 5D space where both there colour and location in the image are taken into account. Thus the clustering will be in the RGBXY hyperspace.

```python
# Create an array of 2D points corresponding to the xy position of each pixel in the image
xy_pts = np.indices(img.shape[:2]).reshape(2, -1).transpose()

# Since colour and position have different units we should weigh them appropriately
# such that both are equally important during clustering. The value of 0.2 is
# empirically determined, but you should try other values e.g. 0.1, 0.5, 1.0
# If the data is first normalised this step will not be necessary.
xy_weight = 0.2
# Stack the 3D RGB points and the 2D xy points into 5D points
rgbxy_pts = np.float32(np.hstack((rgb_pts, xy_weight * xy_pts)))

# Set the number of clusters
K = 16
# Create the criteria when the K-means iterations should be terminated. 
# The iteration will stop if the error is less then EPS = 1.0 or the
# maximum number of iteration MAX_ITER = 20 is achieved
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)

# Run K-means. The return values are:
# ret     - the final value of the minimised sum of distances
# labels  - the cluster which each point belons to
# centers - the centers of the K clusters
ret,labels,centers=cv2.kmeans(rgbxy_pts,  # Input points
                              K,          # Number of clusters  
                              None,       # Initial estimate of the label of each point
                              criteria,   # Termination criteria
                              10,         # Return the best result after that many attempts 
                              cv2.KMEANS_RANDOM_CENTERS)  # Initialise the clusters randomly

# Convert the centers from float32 to uint8 and take only the RGB coordinates
centers = np.uint8(centers[:, :3])

# Set the coordinates of each point to the center of the cluster it belongs to
seg_rgbxy_pts = centers[labels.flatten()]

# Make the RGB points array into an image again
seg_rgbxy_img = seg_rgbxy_pts.reshape((img.shape))

# Show the RGBXY and the RGB segmented image
plt.subplot(1, 2, 1)
plt.title("RGBXY space")
display(seg_rgbxy_img)
plt.subplot(1, 2, 2)
plt.title("RGB space")
display(seg_rgb_img)
```

## Meanshift Segmentation

```python
xy_r = 20 # The spatial window radius
rgb_r = 80 # The color window radius

# Run meanshift
meanshift_img = cv2.pyrMeanShiftFiltering(img, xy_r, rgb_r)

# Display the segmented and the original images
plt.subplot(1, 2, 1)
display(img)
plt.subplot(1, 2, 2)
display(meanshift_img)
```

##GrabCut Segmentation

```python
# Manually set the roi where the foreground of the image is
fg_roi = (250, 550, 380, 500)

# Display the foreground ROI
x, y, w, h = fg_roi
display(cv2.rectangle(img.copy(), (x, y), (x + w, y + h), (0, 255, 255), 3)) 
```

```python
# Create an empty mask where the output of the algorithm will be stored
mask = np.zeros(img.shape[:2], np.uint8)

# Background and foreground are modelled as a Gaussian Mixutre Model over 
# the RGB space. The default number of mixture components is 5 and 13 parameters 
# per component are required (3 mean + 9 covariance + 1 component weight).
# Thus 5 * 13 = 65 parameters are needed for the model. 

# Create an empty array of 1x65 for the background model
bg_model = np.zeros((1, 65), np.float64)
# Create an empty array of 1x65 for the foreground model
fg_model = np.zeros((1, 65), np.float64)

# Apply the GrabCut algorithm
cv2.grabCut(img,      # Input image
            mask,     # Output mask
            fg_roi,   # Foreground ROI
            bg_model, # Empty model of the background
            fg_model, # Empty model of the foreground
            5,        # Number of iterations
            cv2.GC_INIT_WITH_RECT) # Everything outside the foreground ROI is considered background

# Set all pixels in the maske which are foreground (GC_FGC) or 
# probably foreground (GC_PR_FGD) to 1 and all other pixels to 0.
mask = np.where((mask == cv2.GC_FGD) | (mask == cv2.GC_PR_FGD), 1, 0).astype('uint8')

# Apply the mask to the input image
grab_cut_img = img * mask[:, :, np.newaxis]

# Show the segmented foreground
display(grab_cut_img)
```

Better results can be obtained if a more accurate mask is available.

```python
# Load a more accurate mask. 
mask_img = cv2.imread("./imgs/segmentation/mask.png")

# Convert the mask to greyscale
mask = cv2.cvtColor(mask_img, cv2.COLOR_BGR2GRAY)
# Black pixels are background
mask[mask == 0] = cv2.GC_BGD
# Grey pixels are probably background
mask[mask == 128] = cv2.GC_PR_BGD
# White pixels are foreground
mask[mask == 255] = cv2.GC_FGD

# Initialise the foreground and background models
bg_model = np.zeros((1, 65), np.float64)
fg_model = np.zeros((1, 65), np.float64)
# Apply the GrabCut algorithm
cv2.grabCut(img,      # Input image
            mask,     # Input and output mask
            None,     # No foreground ROI
            bg_model, # Empty model of the background
            fg_model, # Empty model of the foreground
            5,        # Number of iterations
            cv2.GC_INIT_WITH_MASK) # Use the provied mask 

# Set all pixels in the mask which are foreground (GC_FGC) or 
# probably foreground (GC_PR_FGD) to 1 and all other pixels to 0.
mask = np.where((mask == cv2.GC_FGD) | (mask == cv2.GC_PR_FGD), 1, 0).astype('uint8')

# Apply the mask to the input image
grab_cut_img = img * mask[:, :, np.newaxis]

# Show the segmented foreground
plt.subplot(1, 2, 1)
display(mask_img)
plt.subplot(1, 2, 2)
display(grab_cut_img)
```
