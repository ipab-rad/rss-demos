# Keypoints & Feature Descriptors Matching

As before we start with setting up the environment and a simple helper function to display images.

```python
# Import matplotlib to display images
import matplotlib.pyplot as plt
%matplotlib inline
# Set the default size of the figures to be 
# width 18.0 and height 13.0 units.
plt.rcParams['figure.figsize'] = (18.0, 13.0)

# Import NumPy in order to execute array operations efficiently.
import numpy as np

# Enable the OpenCV 3 module. The module name is not changed to 
# cv3 yet which could be quite confusing sometimes.
import cv2

# This is a helper function which displays an OpenCV image
def display(img):
    # OpenCV stores images in B G R sequence, so we need
    # to explicitly convert it to R G B in order to plot
    # with matplotlib
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
```

## Keypoints - Harris Corners

```python
scene = cv2.resize(cv2.imread("./imgs/matching/scene.jpg"), (0, 0), fx = 0.5, fy=0.5)
display(scene)
```

```python
box = cv2.resize(cv2.imread("./imgs/matching/object.jpg"), (0, 0), fx = 0.5, fy = 0.5)
display(box)
```

```python
gray = cv2.cvtColor(box, cv2.COLOR_BGR2GRAY)
np.float32(gray)
harris_response = cv2.cornerHarris(gray, 2, 3, 0.04)
thresh = 0.01 * harris_response.max()
corners = np.where(harris_response > thresh)

corners_image = box.copy()
for x, y in zip(corners[1], corners[0]):
    cv2.circle(corners_image, (x, y), 2, (0, 0, 255), -1) 

display(corners_image)
```

## Keypoints Matching

```python
# Create ORB feature detector object
orb = cv2.ORB_create(nfeatures = 1000) # Set the maximum number of returned 
                                       # feature points to 1000

# Create a brute force matcher object.
bf = cv2.BFMatcher(normType=cv2.NORM_HAMMING2, crossCheck=True)

# Detect the feature points in the first two images
box_kp, box_des = orb.detectAndCompute(box, None)
scene_kp, scene_des = orb.detectAndCompute(scene, None)

# Find the matching points using brute force
matches = bf.match(box_des, scene_des)

# Sort the matches in the order of the distance between the 
# matched descriptors. Shorter distance means better match.
matches = sorted(matches, key = lambda m: m.distance)

# Use only the best 1/10th of matchess
matches = matches[:len(matches) / 30]
print(len(box_des))
print(len(scene_des))
print(len(matches))

# Visualise the detected matchin pairs
match_vis = cv2.drawMatches(box, # First processed image
                            box_kp,    # Keypoints in the first image
                            scene,     # Second processed image
                            scene_kp,  # Keypoints in the second image
                            matches,   # Detected matches
                            None)      # Optional output image



scene_pts = np.float32([scene_kp[m.trainIdx].pt for m in matches])
box_pts = np.float32([box_kp[m.queryIdx].pt for m in matches])

# Calculate the homography between the two images. The function works 
# by optimising the projection error between the two sets of points.
H, _ = cv2.findHomography(box_pts, scene_pts, cv2.RANSAC, 2)

h, w = box.shape[:2]
# print(box.shape[:2])
bounds = np.float32([[[0, 0], [w, 0], [h, 0], [w, h]]])

    
# Project the corners in the first image plane
bounds = cv2.perspectiveTransform(bounds, H).reshape(-1, 2)
print(bounds)

img = scene.copy()
match_vis = cv2.circle(match_vis, (int(w + bounds[0][0]), bounds[0][1]), 10, (255, 0, 0), -1)
match_vis = cv2.circle(match_vis, (int(w + bounds[1][0]), bounds[1][1]), 10, (0, 255, 0), -1)
match_vis = cv2.circle(match_vis, (int(w + bounds[2][0]), bounds[2][1]), 10, (0, 0, 255), -1)
match_vis = cv2.circle(match_vis, (int(w + bounds[3][0]), bounds[3][1]), 10, (0, 255, 255), -1)
# cv2.line(img, (bounds[0][0], bounds[0][1]), (bounds[1][0], bounds[1][1]), (255, 0, 0), 10)
# cv2.line(img, (bounds[1][0], bounds[1][1]), (bounds[2][0], bounds[2][1]), (0, 255, 0), 10)
# cv2.line(img, (bounds[2][0], bounds[2][1]), (bounds[3][0], bounds[3][1]), (0, 0, 255), 10)
# cv2.line(img, (bounds[3][0], bounds[3][1]), (bounds[0][0], bounds[0][1]), (255, 255, 255), 10)
# Display the image

display(match_vis)
```
