```python
# Import matplotlib to display images
import matplotlib.pyplot as plt
%matplotlib inline
# Set the default size of the figures to be 
# width 15.0 and height 13.0 units.
plt.rcParams['figure.figsize'] = (15.0, 13.0)

# Import NumPy in order to execute array operations efficiently.
import numpy as np

# Enable the OpenCV 3 module. The module name is not changed to 
# cv3 yet which could be quite confusing sometimes.
import cv2

# This is a helper function which displays an OpenCV image
def display(img):
    # OpenCV stores images in B G R sequence, so we need
    # to explicitly convert it to R G B in order to plot
    # with matplotlib
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
```

You can easily read images from files with OpenCV. Color images are stored as 3-dimensional arrays [height] x [width] x [colors].

```python
img = cv2.imread("imgs/background.png")
# Show the image
display(img)
print "Image array size [H x W x C]: "
print(img.shape)
```

Once loaded, you can perform various functions on an image. Let's calculate the colour histogram of the image.

```python
# Create the bins for each colour channel and set each 
# bin to 0. We would like to have 64 bins per colour
r_bins = np.zeros(64)
g_bins = np.zeros(64)
b_bins = np.zeros(64)


# Iterate over each pixel in the image. 
for i in range(0, img.shape[0]):     # First loop over each row
    for j in range(0, img.shape[1]): # Then loop over each column
        # Get the color of the pixel
        pixel = img[i, j, :]   # the BGR values of the pixel.
        
        # Get each color separately. Remeber that OpenCV stores
        # images in BGR format, not RGB.
        r = pixel[2]           # the R value of the pixel
        g = pixel[1]           # the G value of the pixel
        b = pixel[0]           # the B value of the pixel
        
        # Accumulate the count of each colour value. The maximum
        # colour value is 255 and we have 64 bins so we need to 
        # divide the colour by 4 in order to get a bin index.
        r_bins[r / 4] += 1
        g_bins[g / 4] += 1
        b_bins[b / 4] += 1
        
# Plot the histograms for each colour
plt.plot(r_bins, 'r')
plt.plot(g_bins, 'g')
plt.plot(b_bins, 'b')
plt.xlim([0,64])
plt.show()
```

The prevoious example was useful to demonstrate simple pixel operations, however in practice you should use functions provided by OpenCV. You can get the colour histogram of an image with a single OpenCV function:

```python
# calcHist is a function provided by OpenCV 
r_bins = cv2.calcHist([img],    # a list of the images to be process
                      [2],      # index of channel to be process
                      None,     # mask, we do not use any
                      [64],     # number of bins in the histogram
                      [0, 256]) # range of the histogram
g_bins = cv2.calcHist([img], [1], None, [64], [0, 256])
b_bins = cv2.calcHist([img], [0], None, [64], [0, 256])

# Plot the histograms for each colour
plt.plot(r_bins, 'r')
plt.plot(g_bins, 'g')
plt.plot(b_bins, 'b')
plt.xlim([0,64])
plt.show()
```

Let's load an image of the same background, but with a red object in the front.

```python
obj_img = cv2.imread("imgs/red_object.png")
display(obj_img)
```

Now let's calculate the histogram of the new image with an object and compare it with the background image histogram.

```python
# Get the histograms for each channel
obj_r_bins = cv2.calcHist([obj_img], [2], None, [64], [0, 256])
obj_g_bins = cv2.calcHist([obj_img], [1], None, [64], [0, 256])
obj_b_bins = cv2.calcHist([obj_img], [0], None, [64], [0, 256])

# Plot the old histogram and the new one in a 3x2 grid
plt.subplot(3, 2, 1)
plt.title("Red Object")
plt.plot(obj_r_bins, 'r')
plt.ylim([0,180000])
plt.xlim([0,64])

plt.subplot(3, 2, 2)
plt.title("Only Background")
plt.plot(r_bins, 'r')
plt.ylim([0,180000])
plt.xlim([0,64])

plt.subplot(3, 2, 3)
plt.plot(obj_g_bins, 'g')
plt.ylim([0,180000])
plt.xlim([0,64])

plt.subplot(3, 2, 4)
plt.plot(g_bins, 'g')
plt.ylim([0,180000])
plt.xlim([0,64])

plt.subplot(3, 2, 5)
plt.plot(obj_b_bins, 'b')
plt.ylim([0,180000])
plt.xlim([0,64])

plt.subplot(3, 2, 6)
plt.plot(b_bins, 'b')
plt.ylim([0,180000])
plt.xlim([0,64])

plt.show()
```

As expected there is a distinct increase in the number of red pixels. The distribution of green pixels is more or less the same. The number of bright blue pixels has decreased and the number of dark blue pixels has increased.
