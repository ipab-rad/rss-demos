## Single Camera Calibration

The camera calibration module in OpenCV works by analysing feature points with 3D coordinates known a priori. The camera parameters are estimated by projecting the 3D points on the image plane and minimising the error between the projections and the actually observed feature points. Typically, a chessboard pattern is used because it is easy to detect the corners of the black and white squares and it is easy to describe the 3D structure of the feature points. In order to obtain enough data for accurate calibration multiple views of the pattern are needed.

### Chessboard Initialisation

The chessboard in the images has 10x7 squares which results in 9x6 corners. We assume that the size of the square is one.

```python
# Set the size of the chessboard
board_h = 9
board_w = 6

# Initialise an array of 9x6 3D points which are (0, 0, 0)
board_pts = np.zeros((board_w * board_h, 3), np.float32)

# Set the x and y coordinate of each point to be the 
# row and column of the chessboard corner. z is left to 0.
board_pts[:, :2] = np.mgrid[0:board_h, 0:board_w].T.reshape(-1, 2)

# You can print the array to see the 3D points describing the chessboard
# print(board_pts)
```

### Chessboard Corners

OpenCV provides a function to detect the chessboard corners.

```python
# Copy an image from the loaded ones
img = left_images[2].copy()

# Find the chessboard corners in the image. The function
# returns a boolean indicating whether the entire board 
# was found and an array with the detected corners
found, corners = cv2.findChessboardCorners(img,                 # image to be processed
                                           (board_h, board_w))  # size of the chessboard
# Draw the detected chessboard
img = cv2.drawChessboardCorners(img,                 # image where to draw the cornerss
                                (board_h, board_w),  # size of the chessboard
                                corners,             # the detected cornes
                                found)               # flag indicating if the board was found

# Display the image with the detected corners
display(img)
```

### Calibration

OpenCV provides a function for single camera calibration which requires an array of 3D points and their observed locations in the image. The function estimates both the internal and the external parameters of the camera.

```python
# Initialise two empty arrays
obj_pts = [] # 3D points in real world scene
img_pts = [] # 2D points in image plane

# For each image taken by the left camera
for img in left_images:
    # Detect the chessboard corners
    found, corners = cv2.findChessboardCorners(img, (board_h, board_w))
    
    # Add the precalculated 3D board points
    obj_pts.append(board_pts)
    
    # Add the observed 2D points in the image
    img_pts.append(corners)

# Call the calibrateCamera function which optimises the
# reprojection error. The function returns:
#     retval - the final reprojection error
#     mtx    - the estimated camera projection matrix
#     dist   - the estimated distortion coefficients
#     rvecs  - the orientations of the camera with respect
#              to the chessboard for each image
#     tvecs  - the position of the camera with respect 
#            - to the chessboard for each image
retval, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(obj_pts,       # The 3D points
                                                      img_pts,       # The observed 2D projection on the image
                                                      img.shape[:2], # Size of the image
                                                      None,          # Initial camera matrix
                                                      None)          # Initial distortion coefficients

# Print the estimated intrinsic parameters
print("Reprojection error: ")
print(retval)
print("\n")

print("Camera intrinsic matrix: ")
print(mtx)
print("\n")

print("Radial distortion coefficients: ")
print([dist.ravel()[0],dist.ravel()[1],dist.ravel()[4]])
print("\n")

print("Tangential distortion coefficients: ")
print([dist.ravel()[2],dist.ravel()[3]])
print("\n")
```

### Undistort Images

Once the intrinsic parameters of the camera are estimated we can compensate for lens distortion. The code bellow draws a straight line on the original image and on the undistorted one.

```python
# Show a stright line on the original image
plt.subplot(1, 2, 1)
# Copy one of the loaded images
img = left_images[2].copy()
# Detect the corners
found, corners = cv2.findChessboardCorners(img, (board_h, board_w))
# Draw a line along the first row of corners on the chessboard
img = cv2.line(img, tuple(corners[0].ravel()), tuple(corners[8].ravel()), (255, 0, 255), 2)
# Display the image
display(img)

# Show a stright line on the undistorted image
plt.subplot(1, 2, 2)
# Undistort one of the loaded images
img  = cv2.undistort(left_images[2], # the image to be undistorted
                     mtx,            # the camera matrix
                     dist)           # the distortion coefficients
# Detect the corners in the undistorted image
found, corners = cv2.findChessboardCorners(img, (board_h, board_w))
# Draw a line along the first row of corners on the chessboard
img = cv2.line(img, tuple(corners[0].ravel()), tuple(corners[8].ravel()), (255, 0, 255), 2)
# Display the image
display(img)
```

### Project 3D Points

Given the camera intrinsic and extrinsic parameters we can project 3D points to the image.

```python
# Copy one of the second image
img = left_images[2].copy()

# Detect the corners in the image
found, corners = cv2.findChessboardCorners(img, (board_h, board_w))

# Create three points along the 3D axes
axes = np.float32([[3,0,0], [0,3,0], [0,0,-3]]).reshape(-1,3)

# Project the 3D points to the image plane
img_pts, _ = cv2.projectPoints(axes,      # 3D points to be projected 
                               rvecs[2],  # Orientation of the camera in the second image
                               tvecs[2],  # Position of the camera in the second image
                               mtx,       # Camera matrix
                               dist)      # Distortion coefficients

# Draw lines between the (0, 0, 0) corner and the projected points
origin = tuple(corners[0].ravel())
img = cv2.line(img,                        # image where to draw the line
               origin,                     # first point of the line
               tuple(img_pts[0].ravel()),  # second point of the line
               (255,0,0),                  # (B, G, R) color
               3)                          # width of the line
img = cv2.line(img, origin, tuple(img_pts[1].ravel()), (0,255,0), 3)
img = cv2.line(img, origin, tuple(img_pts[2].ravel()), (0,0,255), 3)

# Show the image
display(img)
```

## Fundamental Matrix Estimation

OpenCV has a built in function to estimate the fundamental matrix of a stereo camera given only corresponding pixels. Several algorithms are available. Try each one and observe the resultin epilines.

```python
# Find the corners of the board in the left and the right image
_, left_corners = cv2.findChessboardCorners(left_images[2], (board_h, board_w))
_, right_corners = cv2.findChessboardCorners(right_images[2], (board_h, board_w))

# Calculate the fundamental matrix
F, _ = cv2.findFundamentalMat(left_corners,     # points in the left image
                              right_corners,    # points in the right image
                              cv2.FM_RANSAC)    # use the RANSAC algorithm
                              #cv2.FM_7POINT)   # use 7 point algorithm 
                              #cv2.FM_8POINT)   # use 8 point algorithm

# Find the epilines in the right image of the
# corners in the left image
lines = cv2.computeCorrespondEpilines(left_corners, # feature points in the left image
                                      1,            # the calculated epilines are for 
                                                    # the right camera. (0 means left)
                                      F)            # The estimated fundamental matrix

# Visualise the epilines:

# Copy the left and the right images
left_img = left_images[2].copy()
right_img = right_images[2].copy()

# Select only 3 corners the epilines of which to be visualised
corner_idxs = [10, 46, 40]

# Select 3 bright colors to aid visability
colors = [(233, 199, 24), 
          (25, 180, 37), 
          (0, 195, 249)]

# Get the size of the image
h,w = left_img.shape[:2]
# for each corner which is to be visualised and its color
for i, clr in zip(corner_idxs, colors):
    # Get the coefficients of the i-th line
    l = lines.squeeze()[i]
    # Calculate the first point segment along the line
    x0,y0 = map(int, [0, -l[2]/l[1]])
    # Calculate the second point segment along the line
    x1,y1 = map(int, [w, -(l[2]+l[0]*w)/l[1]])
    # Draw the line in the right image
    cv2.line(right_img, (x0, y0), (x1, y1), clr, 2)
    # Draw the corresponding feature in the right image
    cv2.circle(right_img, tuple(right_corners[i].ravel()), 10, clr, -1)
    # Draw the feature in the left image
    cv2.circle(left_img, tuple(left_corners[i].ravel()), 10, clr, -1)

# Show the left and the right images    
plt.subplot(1,2,1)
display(left_img)
plt.subplot(1,2,2)
display(right_img)
```

## Stereo Camera Calibration

More often the fundamental matrix is calculated by calibrating the stereo images. Once the stereo camera is calibrated the images can be rectified such that the epilines are horizontal an the corresponding pixels are at the same row in the image.

```python
obj_pts = [] # 3D points in the real world scene
left_pts = [] # 2D points in the left image plane
right_pts = [] # 2D points in the right image plane

# Get the size of the image
img_sz = (left_images[0].shape[1], left_images[0].shape[0]) 

# For each pair of left and right image
for left_img, right_img in zip(left_images, right_images):
    # Detect the corners in the left image
    left_found, left_corners = cv2.findChessboardCorners(left_img, (board_h, board_w))
    # Detect the corners in the right image
    right_found, right_corners = cv2.findChessboardCorners(right_img, (board_h, board_w))
    
    # Add the precalculated 3D board points
    obj_pts.append(board_pts)
    
    # Add the observed 2D points in the left image
    left_pts.append(left_corners)
    
    # Add the observed 2D points in the right image
    right_pts.append(right_corners)

# Calibrate the left and the right cameras individually as shown before
retval, left_mtx, left_dist, _, _ = cv2.calibrateCamera(obj_pts, left_pts, img_sz, None, None)
retval, right_mtx, right_dist, _, _ = cv2.calibrateCamera(obj_pts, right_pts, img_sz, None, None)

# Call the stereoCalibrate function which minimises the reprojection error 
# of the scene 3D points to the left and right image planes. The function returns:
#     retval      - the final reprojection error
#     left_mtx    - the final left camera matrix
#     left_dist   - the final left camera distortion coefficients
#     right_mtx   - the final right camera matrix
#     right_dist  - the final right camera distortion coefficients
#     R           - rotation between the left and the right camera
#     T           - translation between the left and the right camera
#     E           - essential matrix of the stereo camera
#     F           - fundamental matrix of the stereo camera
(retval, 
 left_mtx, left_dist, 
 right_mtx, right_dist, 
 R, T, E, F) = cv2.stereoCalibrate(obj_pts,      # 3D scene points
                                   left_pts,     # points in the left image
                                   right_pts,    # points in the right image
                                   left_mtx,     # left camera matrix
                                   left_dist,    # left camera distortion coefficients
                                   right_mtx,    # right camera matrix
                                   right_dist,   # right camera distortion coefficients
                                   img_sz)       # the size of the analysed images

# Calculate the rectification transform for the stereo camera. 
# left_r and right_r are the rotations of the left and the right
# camera needed to align the epilines.
left_r, right_r, _, _, _, _, _ = cv2.stereoRectify(left_mtx,      # left camera matrix
                                                   left_dist,     # left camera distortion coefficients
                                                   right_mtx,     # right camera matrix
                                                   right_dist,    # right camera distortion coefficients
                                                   img_sz,        # the size of the analysed images
                                                   R,             # rotation between the left and the right camera
                                                   T)             # translation between the left and the right camera

# Initialise a map which rectifies and undistorts the image 
# for the left and the right camera
left_mapx, left_mapy = cv2.initUndistortRectifyMap(left_mtx,      # left camera matrix
                                                   left_dist,     # left camera distortion coefficients
                                                   left_r,        # left camera rotation needed for rectification 
                                                   None,          # we do not need the updated camera matrix
                                                   img_sz,        # the size of the transformed images
                                                   cv2.CV_32FC1)  # use 32-bit floats for the mapping

right_mapx,right_mapy = cv2.initUndistortRectifyMap(right_mtx,right_dist,right_r,None,img_sz, cv2.CV_32FC1)

# Apply the the transform to the left and the right image
# and put them one next to the other
img = np.hstack((cv2.remap(left_images[2],left_mapx,left_mapy,cv2.INTER_LINEAR),
                 cv2.remap(right_images[2],right_mapx,right_mapy,cv2.INTER_LINEAR)))

# Draw horizontal lines across both images. Since the images are 
# undistorted and rectified the lines pass through corresponding pixels.
for i in range(1, 480, 40):
    img = cv2.line(img, (0, i), (img.shape[1], i), (i, 255, 255 - i), 1)

# Show the resulting image
display(img)
```
