As before we start with setting up the environment and a simple helper function to display images.

```python
# Import matplotlib to display images
import matplotlib.pyplot as plt
%matplotlib inline
# Set the default size of the figures to be 
# width 18.0 and height 13.0 units.
plt.rcParams['figure.figsize'] = (18.0, 13.0)

# Import NumPy in order to execute array operations efficiently.
import numpy as np

# Enable the OpenCV 3 module. The module name is not changed to 
# cv3 yet which could be quite confusing sometimes.
import cv2

# This is a helper function which displays an OpenCV image
def display(img, colorbar = False):
    # OpenCV stores images in B G R sequence, so we need
    # to explicitly convert it to R G B in order to plot
    # with matplotlib
    if len(img.shape) == 3:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(img)
    else:
        img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
        plt.imshow(img, cmap='gray')
    if(colorbar):
        plt.colorbar()
```

In this demo we will develop a car classifier based on the bag of words technique. First, we need to load the images which we will use for the demo:

## Car Classification

Now that we have the images loaded we should find the dictionary of features that we are going to use for classification. Executing this cell may take a while!

```python
# Set the number of words for the dictionary
words_count = 20

# Create a bag of words trainer which uses K-Means clustering
bow_trainer = cv2.BOWKMeansTrainer(words_count,   # desired number of clusters (words)
                                   attempts = 10) # number of clustering attempts from 
                                                  # which the best one is selected
# Create an ORB feature detector
orb = cv2.ORB_create(nfeatures = 50,    # maximum number of features returned per image
                     edgeThreshold = 1, # size of the border where features are not detected
                     patchSize = 11)    # size of the patch used to calculate the descriptor

# Add the descriptors for all car images to the bag of words trainer
for img in car_images:
    _, desc = orb.detectAndCompute(img, None)
    bow_trainer.add(np.float32(desc))

# Add the descriptors for all no car images to the bag of words trainer        
for img in nocar_images:
    _, desc = orb.detectAndCompute(img, None)
    bow_trainer.add(np.float32(desc))

print("Total number of feature points: ")
print(bow_trainer.descriptorsCount())

# Perform clustering in order to find the words
voc = bow_trainer.cluster()

# Print the shape of the vocabulary. Each row should contain a word.
print("Vocabulary size:")
print voc.shape
```

We need to get the bag of words for each training image:

```python
# Create a brute force matcher in order to find the closest word
# to each feature in the training set.
bf = cv2.BFMatcher()

# Create a bag of words image descriptor extractor which return
# the bag of words for a given image. 
bow_desc_extractor = cv2.BOWImgDescriptorExtractor(orb, # keypoints descriptor
                                                   bf)  # brute force matcher
# Set the vocabulary to the one which we have estimated
bow_desc_extractor.setVocabulary(np.uint8(voc))

# Get the bag of words for each car image
car_bows = np.empty([0, words_count]) # row 'i' will contain the bag of words for car image 'i'
for img in car_images:
    # Find the keypoints in the image
    kp, _ = orb.detectAndCompute(img, None)
    # Get the bag of words of the image
    bow = bow_desc_extractor.compute(img, kp)
    # Add it to the rest of bag of words 
    car_bows = np.vstack((car_bows, bow))

print("Car BOWs: ")
print(car_bows.shape)
      
# Get the bag of words for each no car image
nocar_bows = np.empty([0, words_count]) # row 'i' will contain the bag of words for no car image 'i'
for img in nocar_images:
    # Find the keypoints in the image
    kp, _ = orb.detectAndCompute(img, None)
    # Get the bag of words of the image
    bow = bow_desc_extractor.compute(img, kp)
    # Add it to the rest of bag of words 
    nocar_bows= np.vstack((nocar_bows, bow))
    
print("No car BOWs: ")
print(nocar_bows.shape)
```

Before training a classifier on the extracted BOWs let's plot the average BOW for each class:

```python
avg_car_bow = np.sum(car_bows, 0) / car_bows.shape[0]
avg_nocar_bow = np.sum(nocar_bows, 0) / nocar_bows.shape[0]

plt.subplot(1, 2, 1)
plt.bar(np.arange(avg_car_bow.shape[0]), avg_car_bow, color = 'g')
plt.ylim(0, 0.15)
plt.title("Average Car Image BOW")
plt.xlabel("Word")
plt.ylabel("Normalised Frequency of Occurence")
plt.subplot(1, 2, 2)
plt.bar(np.arange(avg_nocar_bow.shape[0]), avg_nocar_bow, color = 'r')
plt.ylim(0, 0.15)
plt.title("Average No Car Image BOW")
plt.xlabel("Word")
plt.ylabel("Normalised Frequency of Occurence")
plt.show()
```

Now we will train a logistic regression classifier based on the extracted BOWs. Since there are some issues with the Python bindings in OpenCV 3.0 related to the LogisticRegression class, we train a simple neural network with no hidden layers.

```python
# We use a standard multilayer perceptron
ann = cv2.ml.ANN_MLP_create()
# The network has only one layer with one neuron, essentially 
# obtaining a Logistic Regression classifier.
ann.setLayerSizes(np.array([words_count,      # Number of inputs
                            1]))              # Number of neurons in the output layer
# Set the activation function of each neuron to be a standard sigmoid function
# of the form f(x) = beta * (1 - exp(-alpha * x) / (1 + exp(-alpha * x))
ann.setActivationFunction(cv2.ml.ANN_MLP_SIGMOID_SYM, # sigmoid activation function
                          1,                          # alpha
                          1)                          # beta
# Train the netrwork using backpropagation
ann.setTrainMethod(cv2.ml.ANN_MLP_BACKPROP)
# Set the termination criteria to be either 1000 iterations or error < 1
ann.setTermCriteria((cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 1000, 1))
# Set the learning momentum
ann.setBackpropMomentumScale(0.2)
# Set the learning rate
ann.setBackpropWeightScale(0.2)

# Create the training data which combines car and no car images
train_data = np.vstack((nocar_bows, car_bows))
# Create the corresponing data labels - 0 for no car image and 1 for car image
train_labels = np.vstack((np.zeros([nocar_bows.shape[0], 1]), 
                          np.ones([car_bows.shape[0], 1])))

# Train the nerual network
ann.train(np.float32(train_data),    # training data
          cv2.ml.ROW_SAMPLE,         # each row repsresents a sample
          np.float32(train_labels))  # labels corresponding to the training data

print("ANN successfully trained: ")
print(ann.isTrained())
```

Now let's test the classifier that we have just trained. The accuracy results that you obtain will vary between 75-85% since the clustering of feature points starts from random initial conditions resulting in different clusters.

```python
# Apply the classifier to each car test image
# Get the car test images BOWs
test_car_bows = np.empty([0, words_count])
for img in test_car_images:
    kp, _ = orb.detectAndCompute(img, None)
    bow = bow_desc_extractor.compute(img, kp)
    test_car_bows = np.vstack((test_car_bows, bow))

# Predict the label of each test car image
_, car_results = ann.predict(np.float32(test_car_bows))

# Apply the classifier to each no car test image
# Get the no car test images BOWs
test_nocar_bows = np.empty([0, words_count])
for img in test_nocar_images:
    kp, _ = orb.detectAndCompute(img, None)
    bow = bow_desc_extractor.compute(img, kp).flatten()
    test_nocar_bows = np.vstack((test_nocar_bows, bow))
    
# Predict the label of each test no car image
_, nocar_results = ann.predict(np.float32(test_nocar_bows))


# Calculate the true postive rate of the classifier
true_positive = sum(car_results > 0.5) / float(len(car_results))
print "True Positive Rate: "
print true_positive

# Calculate the true negative rate of the classifier
true_negative = sum(nocar_results < 0.5) / float(len(nocar_results))
print "True Negative Rate: "
print true_negative

# Calculate the overall accuracy of the classifier
accuracy = (true_positive + true_negative) / 2
print "Accuracy: "
print accuracy
```

In the above calculations we used an arbitrary threshold in order to estimate the accuracy of the classifier. A better approach is to evaluate the classifier on its precision-recall curve.

```python
# Get a list of threshold values [0:0.01:1]
thresholds = [0.01 * x for x in range(0, 101)]
# Create an empty vector to store the estimated precision for each threshold
precision = np.zeros((len(thresholds)))
# Create an empty vector to store the estimated recall for each threshold
recall = np.zeros((len(thresholds)))

# For each threshold
for i, thresh in enumerate(thresholds):
    # Count the true positive predictions
    tp = float(sum(car_results > thresh))
    
    # Count the false positive predictions
    fp = float(sum(nocar_results > thresh))
    
    # Count the false negative predictions
    fn = float(sum(car_results < thresh))
    
    # Calculate precision
    precision[i] =  tp / (tp + fp)
    # Calculate recall
    recall[i] = tp / (tp + fn)

# Plot the precision-recall curve
plt.plot(recall, precision)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision - Recall Curve of the Classifier")
plt.show()
    
```

## Car Detection

Now that we have a car classifier we can slide a window along a large scene image in order detect a car within it. First load several images of scenes with cars:

```python
# The path where scene images are stored
path = "./imgs/classification/detection_images/"

# Get the names of all files in the folder
files = []
for (dirpath, dirnames, filenames) in walk(path):
    files.extend(filenames)
    break

# Read each image and add it to the array
scene_images = [cv2.imread(path + f) for f in files]

print("Number of loaded images:")
print(len(scene_images))
```

Create a function which detects the car within an image and call it for each image. The results that you observe depend on the accuracy of the classifier. Accuracy between 80-85% usually results in correct car detections in all images.

```python
# Returns an image with a green rectange drawn around the detected car
# scene - input image
def detect_car(scene):
    (w, h) = (100, 40) # size of the sliding window. It is equal to the size of the training images.
    # A dictionary which will contain all candidate detections
    candidates = {} 
    # Slide the window along the image with 10 pixels step
    step = 10
    for i in range(0, scene.shape[0] - h, step):
        for j in range(0, scene.shape[1] - w, step):
            # Crop the sliding window from the large scene image
            img = scene[i:i+h,j:j+w]
            # Get the keypoints in the window
            kp, _ = orb.detectAndCompute(img, None)
            # Get the bow of the window
            bow = bow_desc_extractor.compute(img, kp)
            # Classify the window
            _, result = ann.predict(np.float32(bow))
            # If the classification is positive
            if result[0][0] > 0.5:
                # Add it to the dictionary of possible candidates with the window coordinates as a key
                candidates[(i, j)] = result[0][0]
    # Get the coordinates of the candindate with strongest response
    (y, x) = max(candidates, key=candidates.get)
    # Copy the scene image as we do not want to modify it
    detection = scene.copy()
    # Draw a green rectangle
    cv2.rectangle(detection,      # drawing image
                  (x, y),         # top left corner
                  (x + w, y + h), # bottom right corner
                  (0, 255, 0))    # color (BGR)
    # Return the image with marked detection
    return detection
    
    

# Create a large figure to draw all images
plt.figure(figsize = (20, 30))

# Detect the car in each scene image
for i in range(len(scene_images)):
    plt.subplot(5, 2, i + 1)
    display(detect_car(scene_images[i]))
```
